```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../figure/lena_color_256.tif', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ker = np.array([[ 0, 1, 0], [1, -4, 1], [0, 1, 0.0]])
img1 = cv2.filter2D(img, -1, np.array(ker))
fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(img)
ax2.imshow(img1)
ax1.axes.xaxis.set_visible(False)
ax1.axes.yaxis.set_visible(False)
ax2.axes.xaxis.set_visible(False)
ax2.axes.yaxis.set_visible(False)
ax1.set_title("Original Image")
ax2.set_title("Filtered Image")
# plt.savefig('../figure/lena_filtered_image.png', dpi=300, bbox_inches='tight')
```

```python
# Pooling
img2 = np.zeros((img.shape[0]//2, img.shape[1]//2, img.shape[2]), dtype=np.uint8)
H, W, C = img1.shape
for h in np.arange(0, H//2):
    for w in np.arange(0, W//2): 
        img2[h, w, :] = np.max(np.max(img1[(2*h):(2*h+2), (2*w):(2*w+2), :], axis=0), axis=0)
```
