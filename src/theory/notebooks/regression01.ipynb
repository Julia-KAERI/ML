```julia
using LinearAlgebra, CairoMakie, NAJ, LaTeXStrings
```

```julia
xs = 1:0.5:10
f(x, p) = p[1]  + p[2] * x^1 +p[3] * x^2 + p[4] * x^3
p0 = [-9.0, 11.8, -3.0, 0.2]
f(x) = f(x, p0)
ys = f.(xs) .+ 1.0*(rand(length(xs)) .- 0.5);
Φ = [x^k for x in xs, k in 0:3]
θ = inv(Φ'*Φ) * Φ' * ys
f1(x) = f(x, θ)
```

```julia
fig = Figure()
ax = Axis(fig[1,1], xtickalign = 1.0, ytickalign=1.0)
scatter!(ax, xs, ys, markersize=10)
lines!(ax, xs, f1.(xs), color = "red")
#save("polyreg.png", fig)
fig
```

```julia
xs = 0:1:10
f(x, p) = sum([p[j]x^(j-1) for j in 1:length(p)])
p0 = [-9.0, 11.8, -3.0, 0.2, -1.2]
f1(x) = f(x, p0)
ys = 3.0 .* xs .* sin.(xs) .+ 0.3 .* (rand(length(xs)) .- 0.5)
Φ1 = [x^k for x in xs, k in 0:14]
θ1 = inv(Φ1'*Φ1) * Φ1' * ys
f2(x) = f(x, θ1)
```

```julia
fig = Figure()
ax = Axis(fig[1,1], xtickalign = 1.0, ytickalign=1.0)
scatter!(ax, xs, ys, markersize=12)
st = 0.0:0.01:10.0
lines!(ax, st, f2.(st), color = "red")
ylims!(ax, [-20, 30])
#save("polyreg.png", fig)
fig
```

```julia
xs = [-4, 7, -5, -9, 2, -6, 1.0, 10, 5];
ys = [13, -3, -16, -19, -12, 1, 9, -5, 0];
f(x, p) = sum([p[j]x^(j-1) for j in 1:length(p)])
fig = Figure()
ax = Axis(fig[1,1], xtickalign = 1.0, ytickalign=1.0)
scatter!(ax, xs, ys, markersize=12)
st = -9:0.01:10.1
Φ1 = [x^k for x in xs, k in 0:2]
θ1 = inv(Φ1'*Φ1) * Φ1' * ys
f1(x) = f(x, θ1)
lines!(ax, st, f1.(st), color = "red", label = "2nd")

Φ1 = [x^k for x in xs, k in 0:4]
θ1 = inv(Φ1'*Φ1) * Φ1' * ys
f1(x) = f(x, θ1)
lines!(ax, st, f1.(st), color = "green", label = "4th")


Φ1 = [x^k for x in xs, k in 0:5]
θ1 = inv(Φ1'*Φ1) * Φ1' * ys
f1(x) = f(x, θ1)
lines!(ax, st, f1.(st), color = "violet", label = "5th")


Φ1 = [x^k for x in xs, k in 0:8]
θ1 = inv(Φ1'*Φ1) * Φ1' * ys
f1(x) = f(x, θ1)
lines!(ax, st, f1.(st), color = "black", label = "8th")

# save("overfit_0.png", fig)
ylims!(ax, [-30, 80])
axislegend(ax)
#save("overfit.png", fig)
fig
```
