```julia
using CairoMakie, LaTeXStrings
```

```julia
function mandelbrot(x, y)
    z = c = x + y*im
    for i in 1:30.0; abs(z) > 2 && return i; z = z^2 + c; end; 0
end

f=Figure()
ax = Axis(f[1, 1], aspect = 1.0)

heatmap!(ax, -2:0.001:1, -1.1:0.001:1.1, mandelbrot,
    colormap = Reverse(:deep))
#text!(ax, 0, 0, "A")
f
```

```julia
fs(x) = 0.04*(x-3)*(x-2)*(x-1.5)*(x+1.5)*(x+2)*(x+3)*exp(-x^2/7)
t = collect(-5:0.1:5)

fig1=Figure(size=(600, 400))

ax1 = Axis(fig1[1, 1], limits=(-5, 5, -5, 10) )
x1=-2.6
text!(ax1, L"x_G", position=(0,fs(0)-0.2), align=(:center, :top))
text!(ax1, L"x_L", position=(x1,fs(x1)), align=(:center, :top))
lines!(ax1, t, fs.(t),)


scatter!(ax1, [0.0,],  [fs(0.0), ], color="red")

scatter!(ax1, [x1,],  [fs(x1), ], color="green")

#save("local_global_minimum.png", fig1)
fig1
```
